The program uses the scikitlearn package in Python to perform k-means clustering on the breast cancer dataset. With the help of K-means clustering, data is divided into k bunches, and each information point is sent to the cluster that contains the nearest cruel (centroid). The method begins by stacking the dataset for breast cancer, which has 30 unique characteristics for each perception. Using Central Component Analysis, the dataset's dimensionality is reduced from 30 to 2 to display the information (PCA). PCA is a measurement-diminishing technique that organizes data into a new framework while retaining the most essential information. Then, with n clusters=10, the K-Means lesson from sci-kit-learn is imported. This shows that we are looking for ten clusters of information inside the data. The balanced dataset is then fit and predicted using the k-means approach, which assigns each perception to one of ten clusters based on the similarity of their highlight values. The names variable spares the cluster allocations. Finally, the code utilizes matplotlib. pyplot to produce a 2D diffuse plot of the information foci. Each perception is addressed on a two-dimensional organized framework, with the x-axis referring to the main rule component and the y-axis referring to the present situation. Each of the ten bunches' perceptions has an unexpected color to make them easier to see. We could identify fundamental patterns or structures in the data thanks to how the k-means computation appeared to have built comparative perceptions. The code applies k-means clustering to the breast cancer dataset, lowering the dimensionality for visualization before displaying the clustering outcomes in a 2D scramble plot. This method helps compile the information into information point distribution and identify potential groups or clusters within the breast cancer dataset.

```python
from sklearn.datasets import load_breast_cancer
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

data = load_breast_cancer().data

pca = PCA(2)
df = pca.fit_transform(data)

kmeans = KMeans(n_clusters=10)

label = kmeans.fit_predict(df)

for i in range(10):
    plt.scatter(df[label == i, 0], df[label == i, 1], label=i)

plt.legend()
plt.show()
```

**Part 2 (25 points): Repeat Part I for digits dataset (use load_digits() instead). Use density-based clustering (use DBSCAN(min_samples = 10, eps = 1.5) instead of KMeans(n_clusters = 10)). Draw the plot for labels 0, 1, 2, ..., 22**

The digits dataset is momentarily used in this code, and like recently, PCA is used to reduce its dimensionality to two. This time, we do density-based clustering using the sci-kit-learn DBSCAN course. The DBSCAN calculation bunches attention based on their thickness or a predetermined number of clusters, unlike k-means. It has two essential requirements: Eps of 1.5 and min_samples of 10. We extract the cluster of 0 to 22 names from the labels variable and exhibit each cluster separately after fitting the DBSCAN example to contemporary data. By observing the clusters and how DBSCAN arranges the digits in 2D space, we can observe them. The number of unique names may vary from run to run because DBSCAN organically selects the number of clusters based on information thickness.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.cluster import DBSCAN

digits = load_digits()
X, y = digits.data, digits.target

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

dbscan = DBSCAN(min_samples=10, eps=1.5)
labels = dbscan.fit_predict(X_pca)

unique_labels = np.unique(labels)
n_clusters = len(unique_labels)

for i in range(n_clusters):
    if unique_labels[i] == -1:
        color = 'k'
    else:
        color = plt.cm.tab20(i % 20)
    indices = np.where(labels == unique_labels[i])

    plt.scatter(X_pca[indices, 0], X_pca[indices, 1], c=color, label=f'Label {unique_labels[i]}')

plt.title('DBSCAN Clustering on Digits Dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.show()
```

**Part 3 (25 points): Repeat Part I for iris dataset (use load_iris() instead). Use hierarchical clustering with 5 clusters (use AgglomerativeClustering(n_clusters = 5) instead of KMeans(n_clusters = 10))**

As with earlier examples, we moment the iris dataset in this code and use PCA to reduce its dimensionality to 2. However, we will use Scikit-learn's AgglomerativeClustering course this time to do hierarchical clustering. The agglomerative clustering approach creates a cluster chain of command by consolidating or merging clusters and merging smaller clusters to produce larger ones. We obtain the cluster names from the names variable and show each cluster exclusively after fitting the AgglomerativeClustering model to the revised data. n_clusters alternatively is set to 5. Hierarchical clustering uses information from two measurements to organize the iris. The code applies several leveled clustering on the iris dataset with five clusters, reduces the dimensionality for introduction, and displays the clustering findings in a 2D diffuse plot. The iris dataset's clustering designs can be examined using this method

```python
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt

data = load_iris().data

pca = PCA(2)
df = pca.fit_transform(data)

hierarchical = AgglomerativeClustering(n_clusters=5)

label = hierarchical.fit_predict(df)

for i in range(5):
    plt.scatter(df[label == i, 0], df[label == i, 1], label=i)

plt.legend()
plt.show()
```

**Part-1**

On the digits dataset, we investigated the use of clustering computations with a broader range of clusters (n_clusters = 20). Using Central Component Analysis (PCA), we first reduced the dataset's measurements to two so that we could visualize the data in a 2D coordinate system. To group comparable digit designs for the main piece, we used a K-means clustering method using 20 clusters. The scramble plot displays specific clusters corresponding to various digit classes, successfully allocating the digits into their respective bunches.

```python
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

data = load_digits().data

pca = PCA(2)
df = pca.fit_transform(data)
kmeans = KMeans(n_clusters=20)

label = kmeans.fit_predict(df)

for i in range(20):
    plt.scatter(df[label == i, 0], df[label == i, 1], label=f'Cluster {i}')

plt.legend()
plt.title("K-means Clustering on Digits Dataset (n_clusters = 20)")
plt.show()
```

**part-2**

Using the Agglomerative Clustering calculation with 20 clusters, we thought of hierarchical clustering. When we showed the modified dataset on the 2D plot framework, we noticed a different grouping approach than K-means. Within the PCA-transformed space, progressive clustering created clusters based on the proximity of information focuses, resulting in a progressive tree structure of clusters.

```python
from sklearn.decomposition import PCA
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt

data = load_digits().data
pca = PCA(2)
df = pca.fit_transform(data)

agg_clustering = AgglomerativeClustering(n_clusters=20)

label = agg_clustering.fit_predict(df)
for i in range(20):
    plt.scatter(df[label == i, 0], df[label == i, 1], label=f'Cluster {i}')

plt.legend()
plt.title("Hierarchical Clustering on Digits Dataset (n_clusters = 20)")
plt.show()
```
